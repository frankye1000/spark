```python
from pymongo import MongoClient
client = MongoClient("mongodb://10.120.37.108:27017")
db = client["statistic_data"]
pair_collection = db["pair"]
post_collection = client["project"]["posts"]
plitician_ids = post_collection.find({}).distinct("politician_id").sort({"politician_id":1})
```

```python
plitician_ids = sorted(plitician_ids)
plitician_ids
```

```python
# '朱立倫','盧秀燕','馬英九','陳其邁','柯文哲','姚文智','賴清德','林佳龍','蔡正元','江啟臣','陳菊','侯友宜','鄭文燦','李錫錕','林昶佐','黃國昌','蔡英文','丁守中','蔣萬安','洪慈庸'
```

```python
matrix = []

for i in plitician_ids:
    temp = []
    for j in plitician_ids:
        li = sorted([i, j])
        p = list(pair_collection.find({"pair": li}, {"_id":0}))
        if p:
            temp.append(p[0]["numbers"] )
        else:
            temp.append(0)
    matrix.append(temp)   
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(matrix,index=['朱立倫','盧秀燕','馬英九','陳其邁','柯文哲','姚文智','賴清德','林佳龍','蔡正元','江啟臣','陳菊','侯友宜','鄭文燦','李錫錕','林昶佐','黃國昌','蔡英文','丁守中','蔣萬安','洪慈庸'],columns=['朱立倫','盧秀燕','馬英九','陳其邁','柯文哲','姚文智','賴清德','林佳龍','蔡正元','江啟臣','陳菊','侯友宜','鄭文燦','李錫錕','林昶佐','黃國昌','蔡英文','丁守中','蔣萬安','洪慈庸'])
```
